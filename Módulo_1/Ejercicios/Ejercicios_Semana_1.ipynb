## Ejercicio 3:


*El índice de masa corporal (IMC) se puede calcular utilizando la siguiente fórmula:*

> *IMC = P / A^2*

> *donde P es el peso de la persona en kilogramos y A es la altura de la misma persona en metros. La fórmula resulta en un número que se utiliza para determinar si una persona está en un peso saludable o no.* 

*Un IMC normal se encuentra entre 18.5 y 24.9. Un IMC menor a 18.5 se considera bajo peso, mientras que un IMC mayor a 25 se considera sobrepeso. Un IMC mayor a 30 se considera obesidad. Sin embargo, estos valores son solo una guía y no son válidos para todas las personas, especialmente aquellas con una gran cantidad de músculo o deportistas.*

**Creá un programa que te pida las siguientes informaciones: nombre, edad, peso en kg y altura en m y que te devuelva el nombre, la edad y el índice de masa corporal de esa persona.**
