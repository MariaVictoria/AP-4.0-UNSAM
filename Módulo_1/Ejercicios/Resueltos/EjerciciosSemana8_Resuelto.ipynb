## Ejercicio 3:

Un conjunto de datos (*dataset*) muy usado para ejercitar `Pandas` es el archivo [titanic.csv](https://drive.google.com/file/d/1edbySvT2KUDelJq85csEA6zxiOJ7TvXH/view?usp=share_link), que contiene información sobre los pasajeros del Titanic, con la descripción de las variables que se puede encontrar [aquí](https://drive.google.com/file/d/14ytgtHkAqbcpbUyWQZkNvw-ff7tisuXK/view?usp=share_link). 
Para este dataset hacer un programa que:

1. Carge los datos en un DataFrame y explorarlos
2. Muestre en pantalla las dimensiones, el número de datos que contiene y que muestre los tipos de datos de cada columna. 
3. Muestre en pantalla los datos del pasajero con identificador 148.
4. Encuentre los nombres de las personas que iban en primera clase y los muestre ordenados alfabéticamente.
5. Elimine del DataFrame los pasajeros/as con edad desconocida.
6. Determine la edad media de las mujeres que viajaban en la clase 1.
7. Haga un histograma de la distribución de edades de los sobrevivientes sin distinguir el genero.

````python
df = pd.read_csv("titanic.csv")

# Muestra informacion del DataFrame
df.info()
# Muestra las dimensiones, el número de datos 
print("Dimensiones del DataFrame:", df.shape)
print("Número total de datos:", df.size)
````

````python
# Muestra los datos del pasajero con identificador 148:
print("Datos del pasajero con identificador 148:\n", df[df["IdPasajero"]==148])

# Otra forma:
df[df["IdPasajero"]==148]
````

````python
# Encuentra los nombres de las personas que iban en primera clase y los muestra ordenados alfabéticamente
primera_clase = df[df['Clase'] == 1] #Filtro los datos apra quedarme con los pasajeros que estan el primera clase
nombres = primera_clase['Nombre'].sort_values() #Ordenados alfabéticamente

print("Nombres de las personas que iban en primera clase:\n", nombres)
````

````python
# Elimina del DataFrame los pasajeros/as con edad desconocida
# El metodo .notna() busca los valores que no son nulos (not-null) en un Dataframe o Serie de pandas y devuelve un booleano con True (si no es nulo) o False (si es nulo).
df = df[df['Edad'].notna()]
````

````python
# Determina la edad media de las mujeres que viajaban en la clase 1

mujeres_primera_clase = df[(df['Genero'] == 'female') & (df['Clase'] == 1)]

edad_media =mujeres_primera_clase['Edad'].mean()
print("Edad media de las mujeres que viajaban en primera clase:", edad_media)
````

````python
# Hace un histograma de la distribución de edades de los sobrevivientes sin distinguir el género
sobrevivientes = df[df['Sobrevivencia'] == 1]
sobrevivientes['Edad'].plot.hist(bins=20)
````

````python
# 1)Cargo el dataset
import pandas as pd

df_covid = pd.read_csv('casos_covid_confirmados_santafe.csv')
df_covid.head()
````

````python
# Mostrar el número total de filas y columnas
print("Número de filas: ", len(df_covid ))
print("Número de columnas: ", len(df_covid .columns))
````

````python
# Mostrar que valores toma la columna "clasificacion"
print(df_covid.clasificacion.unique())
````

````python
# 2)Calcular cuánta gente de Santa Fe falleció por covid.
fallecidos = df_covid.fallecido.value_counts() #Valores de la columna fallecidos
print(fallecidos)

100*fallecidos.loc['SI']/(fallecidos.loc['SI']+fallecidos.loc['NO']) #porcentaje que terminaron en muertes.
````

````python
# 3)Calcule cuántos casos y muertes hubo en cada departamento de la provincia de Santa Fe.
df_departamento=df_covid.groupby("residencia_departamento_nombre")["fallecido"] #agrupo por departamento y me quedo con la columna "fallecido"
df_departamento.value_counts() #muestro el conteo de valores de fallecidos
````

````python
# d)
# todos los confirmados
df_covid[(df_covid.edad_años_meses=='Años') & (df_covid.edad<101)].hist(column="edad", bins=20)

# solo fallecidos
df_covid[(df_covid.edad_años_meses=='Años') & (df_covid.edad<101) & (df_covid.fallecido=='SI')].hist(column="edad", bins=20)
````

````python
# 5) gráfico de casos diarios de covid
df_covid['fecha_apertura'] = pd.to_datetime(df_covid['fecha_apertura'], format= '%Y-%m-%d')
df_covid_fechas = df_covid['fecha_apertura'].value_counts().sort_index()
#df_covid.plot.hist(columns="fecha_apertura")
df_covid_fechas.plot.line(figsize=(18,8), grid=True)
````
