## Anatomía de una figura de matplotlib

La anatomía de una figura de matplotlib tiene una jerarquía. Dentro de una figura (figure), puede ir un sistema de ejes (Axes). Un sistema de ejes naturalmente encapsula un plot y tiene ejes X, Y (si es 2D). Y cada eje a su vez contiene sus elementos propios (ticks, etiquetas/labels). 

 * Figura (Figure, plt.figure)
  * Ejes (Axes, plt.subplot())
    * Eje X `===>` xticks
    * Eje Y `===>` yticks

<img src="https://matplotlib.org/stable/_images/anatomy.png" width="600">


```python
import matplotlib.pyplot as plt

# agregamos un lienzo (canvas, figura)
# en matplotlib esto es un objeto de tipo 'figure'
fig = plt.figure()

# agregamos un contenedor dentro de la figura, para graficar algo
# en matplotlib esto es un objeto de tipo 'axes'
ax = fig.add_subplot(1,1,1) # numero de filas, columnas, y numero de plot 

print(type(fig))
print(type(ax))

x = [0.2, 0.4, 0.6, 0.8]
y = [0.4, 1.6, 3.6, 6.4]

ax.plot(x,y)
```

En muchos sitios van a ver que es muy común invocar a matplotlib usando `plt.subplots()` que nos devuelve una tupla `(figure, axes)`. Contar con `figure` y `axes` como objetos separados (desempaquetar la tupla) es útil para después hacer operaciones propias de la figura (ej `fig.savefig("mi_figura.png")`) o propias del sistema de ejes (`ax.set_xlabel="Distancia"). 

```python
# en forma abreviada
fig, ax = plt.subplots()

# trabajar sonbre 
ax.set_xlabel="Distancia"
ax.set_ylabel="Altura"

# 
#fig.savefig("mi_figura.png")
```

## Anotaciones en los gráficos 

Además de graficar datos x,y de variables, podemos adornar y anotar nuestras figuras en matplotlib con texto, flechas, y otros elementos que ayuden a entender el gráfico. 

En general todas las anotaciones y agregados de texto o elementos necesitan que indiquemos en qué posición tiene que aparecer en el gráfico. Para eso se usa el mismo sistema de coordenadas x,y. Y para que sea más intuitivo, en la misma escala que las variables que estamos graficando. 

```python
# inicializamos un plot 
fig, ax = plt.subplots(figsize=(5, 2.7))

# generamos datos x,y
import numpy as np 
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2 * np.pi * x)  # función coseno 

# graficamos 
ax.plot(x, y, linewidth=1)

# y agregamos una anotación 
ax.annotate('máximo local', xy=(2, 1), xytext=(3,1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)

# trazar barras verticales 
ax.vlines(1, color="red", ymin=-1.5, ymax=1.2)
ax.vlines(3, color="pink", ymin=-0.5, ymax=0.5)

# agregar texto 
ax.text(2.5,-1.5, "Esta es la función coseno", color="firebrick")

# agregar rectangulos de colores 
# axvspan = un span vertical en el eje X (an X vertical span)
# por defecto toma todo el eje Y
ax.axvspan(0.5, 1.5, color="green", alpha=0.2)

# axhspan = un span horizontal en el eje X (an X horizontal span)
# por defecto toma todo el eje X
ax.axhspan(-1, 1, color="brown", alpha=0.25)
```

## Agrandar y achicar 

Podemos magnificar (zoom in) o achicar (zoom out) porciones de los gráficos jugando con `margins`. En matplotlib `margins` agrega espacio en blanco (padding) entre los datos y los ejes, efectivamente genera un *marco* (como en el marco de un cuadro). Nos permite *separar* a los puntos o lineas del gráfico de los bordes. Esto podemos aprovecharlo para generar magnificaciones.

```python
# inicializamos un plot 
fig = plt.figure(figsize=(5, 2.7), layout="tight")

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# generamos datos x,y
import numpy as np 
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2 * np.pi * x)  # función coseno 

# graficamos 
ax1.plot(x, y, linewidth=2)
ax1.margins(0) # margins = 0 encaje perfecto en el recuadro
ax1.set_title("Fit")

# zoom out 
ax2.plot(x, y, linewidth=2)
ax2.margins(2,2) # mucho margen en blanco 
ax2.set_title('Zoom out')

# zoom in 
ax3.margins(1,1) # menos margen 
ax3.plot(x, y, linewidth=2)
ax3.set_title('Zoom in')

# zoom in 
ax4.margins(-0.25,-0.25) # valores negativos hacen zoom
ax4.plot(x, y, linewidth=2)
ax4.set_title('Zoom in++')
```

Vamos a hacer un ejercicio usando datos de cotizaciones de acciones. En la bolsa, las acciones tienen códigos cortos (llamados *tickers*) por las cuales las podemos reconocer. 

Vamos a usar además el paquete `yfinance` (Yahoo Finance) que nos permite acceder a datos históricos de cotizaciones. Los datos los recibimos ya en formato de un dataframe de Pandas. Asi que ya sabemos como explorarlos y usarlos. 

El ejercicio consiste en graficar la cotización al cierre, a lo largo del tiempo, de una serie de acciones de interés, y enfocarnos en algunos puntos de interés. A continuación una breve introducción a como usar yfinance y graficar fechas, que hasta ahora no habiamos usado. 



```python
import yfinance as yf # Yahoo Finance
import matplotlib.pyplot as plt
import pandas as pd

# las acciones de Amazon tienen el identificador ('ticker'): "AMZN"

# queremos analizar datos entre 01/01/2000 y 12/31/2016.
# las fechas son YYYY-MM-DD
desde = '2010-01-01'
hasta = '2016-12-31'

# Yahoo Finance nos devuelve un dataframe pandas 
amazon = yf.download(tickers = "AMZN", start = desde, end = hasta)
microsoft = yf.download(tickers = "MSFT", start = desde, end = hasta)

# exploremos los datos 
print(amazon.head())
print(type(amazon))
```

Que datos tenemos? (ver columnas del dataframe):

 * Date: fecha (dia) de la cotización
 * Open: la cotización de apertura de la acción
 * High: la cotización más alto de la acción 
 * Low: la cotización más bajo  de la acción
 * Close: la cotización de la acción al cierre 
 * Adj Close: la cotización ajustado despues de pagar dividendos
 * Volume: el volumen total de acciones negociadas ese día 

```python
# Veamos como se distribuyen los precios al cierre
cierre = amazon['Close']
cierre.describe()
```

```python
fig = plt.figure(figsize=(14,5))
ax = fig.add_subplot()

# ploteamos los datos
# ojo que 'Date' es el indice del DataFrame Pandas, no es una Serie cualquiera!
ax.plot(amazon.index, amazon['Close'], color="tab:brown")
ax.plot(microsoft.index, microsoft['Close'], color="tab:blue")

ax.set_title("Cotización de las acciones de Amazon",size='x-large',color='tab:blue')

# qué paso a fines de Enero y princios de Febrero de 2016?
# https://www.nytimes.com/2016/02/10/technology/fall-in-tech-stocks-is-faster-and-farther-than-broader-market.html
ax.annotate('Caída!', xy=(pd.to_datetime("2016-2-9"),32), xytext=(pd.to_datetime("2016-2-9"), 40),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.vlines(pd.to_datetime("2016-2-9"), color="blue", ymin=5, ymax=30)
```

## Apéndice: usos y Costumbre en matplotlib

En `matplotlib` hay dos maneras principales de acceder a las funciones, métodos y objetos de un plot. La manera actual y recomendada es la que vimos en esta clase. Usar el modo *orientado a objetos*, donde cada objeto (figure, axes, etc) tiene sus métodos y los llamamos o accedemos en forma apropiada `fig.legend()`, `ax.set_xlabel()` etc. En la comunidad matplotlib esto se llama modo *explícito* tiene la ventaja de que nos permite controlar siempre el foco (objeto) a donde van dirigidas las acciones. 

En diversos foros (stackoverflow) y sitios van a ver que hay ejemplos y sugerencias basadas en otro modo de usar matplotlib que no vimos en esta clase y que es el modo *implicito* que va agregando elementos (datos, estilos, texto) al último objeto en uso (activo). Suele ser útil para hacer borradores *rápidos* de gráficos, pero si la figura es compleja, con muchos ejes, etc es complicado seguir el foco. 

Van un par de ejemplos para ilustrar el punto. Ver además: [API Interfaces.](https://matplotlib.org/stable/users/explain/api_interfaces.html)

```python
import numpy as np
import matplotlib.pyplot as plt 

## Object-oriented (OO), Explicit

# data 
x = np.linspace(0, 2, 100)  

# figure
fig = plt.figure(figsize=(5, 2.7), layout='constrained')

# axes
ax = fig.add_subplot(1,1,1)

# plots
lplot = ax.plot(x, x, label='linear')  # Plot some data on the axes.
qplot = ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
cplot = ax.plot(x, x**3, label='cubic')  # ... and some more.

# labels
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.

# legend
ax.legend()  # Add a legend.


print(type(lplot))
print(lplot)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

# pyplot-style, implicito

# data
x = np.linspace(0, 2, 100)  # Sample data.

# figure 
plt.figure(figsize=(5, 2.7), layout='constrained')

# plots 
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')

# labels
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")

# legend
plt.legend()
```
